```python
#Number 1
def is_prime(num):
 if num<2:
    return False
 for i in range(2,int(num**0.5)+1):
  if num % i==0:
        return False
 return True
def display_primes(start,end):
 Prime_number= [num for num in range(start,end+1) if is_prime(num)]
 return Prime_number
start=5
end=50
print(f"Prime numbers between {start} and {end}:")
print(display_primes(start,end))
```

```python
x=df.drop('Loan Category',axis=1)
y=df['Loan Category']
x_train,x_test,y_train,y_test= train_test_split(x,y, test_size=0.30)
```

```python
from sklearn.preprocessing import StandardScaler
scale= StandardScaler()
x_train = scale.fit_transform(x_train)
x_test =  scale.fit_transform(x_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
k=4
n=KNeighborsClassifier(n_neighbors=k)
n.fit(x_train, y_train)
n
```

```python
y_predict = n.predict(x_test)
from sklearn import metrics
print('Training set:', metrics.accuracy_score(y_train,n.predict(x_train)))
print('Testing set:', metrics.accuracy_score(y_test,n.predict(x_test)))
```
